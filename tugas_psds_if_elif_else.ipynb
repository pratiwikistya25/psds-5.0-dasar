#IF ELIF ELSE Stetements

**Pyhton IF ELIF ELSE Statement syntax**


```
if test expression:
  Body of if
elif test expression:
  Body of elif
else:
  Body of else
```

**Python IF ELIF ELSE Flowchart**

IMG_20211015_161741.jpg

**Example: Python IF ELIF ELSE Stetement**


````python
#Program disusun untuk mengecek apakah angka merupakan bilangan positif, negatif, nol
#dan menampilkan tulisan yang sesuai

angka = 3.44

if angka > 0:
  print("Bilangan Positif")
elif angka == 0:
  print("Nol")
else:
  print("Bilangan Negatif")
````

````python
# Statement if
x=4
if x % 2 == 0: # jika x dibagi 2 sisanya samadengan 0
 print("x habis dibagi dua")

#Statement if ... elif ... else
x = 7
if x % 2 == 0: # Jika x dibagi 2 sisanya samadengan 0
 print("x habis dibagi 2")
elif x % 3 ==0: #jika x dibagi 3 sisanya samadengan 0
 print("x habis dibagi tiga")
elif x % 5 == 0: #Jika x dibagi 5 sisanya samadengan 0
 print("x habis dibagi lima")
else:
  print("x tidak habis dibagi dua, tiga ataupun lima")
````

Buatlah sebuah fungsi "Mahal_diskon" dengan ketentuan:


*   input berupa data harga produk dengan tipe data numerik tunggal
* kategori input adalah "Mahal" dan "Murah", kemudian diskon ada 3 kategori"10%", "20%", dan "40%"
* output pertama sebuah teks/string yang menyatakan "mahal" atau "murah" dan keterangan "kamu dapat diskon..."
*Output kedua adalah harga yang dibayarkan.



````python
#ketikan codingan anda

#Harga dalam satuan ribu rupiah
Harga_Produk = int(input("Harga produk :"))
if Harga_Produk >= 100:
  print("Mahal")
  if Harga_Produk <= 1000:
    print("kamu dapat diskon 20%")
    diskon = Harga_Produk*(20/100)
    Harga_diskon = Harga_Produk - diskon
    print("harga yang dibayarkan", Harga_diskon)
  else:
    print("kamu dapat diskon 10%")
    diskon = Harga_Produk*(10/100)
    Harga_diskon = Harga_Produk - diskon
    print("Harga yang dibayarkan :", Harga_diskon)
else:
  print("Murah")
  print("kamu dapat diskon 40%")
  diskon = Harga_Produk*(40/100)
  Harga_diskon = Harga_Produk-diskon
  print("harga yang dibayarkan :", Harga_diskon)
````
